```python
%matplotlib qt
plan_id =[]
duration = []
railing_min = []
railing_max = []
railing_mean = []
exposure_min = []
exposure_max = []
exposure_mean = []
import numpy as np
from natsort import natsorted, ns
all_content = []
with open("res.txt",'r') as f:
    lines = f.readlines()
    for idx,line in enumerate(lines):
        line = line.replace('\n','')
        if idx in [0,1,2]:
            pass
        else:
            content = line.split(',')
            all_content.append(content)



lines = natsorted(all_content, key=lambda x: x[0])
for line in lines:
    plan_id.append(line[0])
    duration.append(float(line[1]))
    railing_min.append(float(line[2]))
    railing_max.append(float(line[3]))
    railing_mean.append(float(line[4]))
    exposure_min.append(float(line[5])*100)
    exposure_max.append(float(line[6])*100)
    exposure_mean.append(float(line[7])*100)
# lines[lines[:, 0].natsort()]
print(lines)
            # print(content)
import matplotlib.pyplot as plt
fig_plan_comp = plt.figure("Plan Comparison")
ax_plan_comp_duration = fig_plan_comp.add_subplot(3, 1, 1)
ax_plan_comp_rail = fig_plan_comp.add_subplot(3, 1, 2)
ax_plan_comp_exposure = fig_plan_comp.add_subplot(3, 1, 3,sharex = ax_plan_comp_rail)
ax_plan_comp_duration.set_title("Construction Plan Comparison")
ax_plan_comp_duration.set_ylabel("Plan Duration [Days]")
ax_plan_comp_rail.set_ylabel("Railing demand [m]")
ax_plan_comp_exposure.set_ylabel("Work Crew Exposure [%]")
# ax_plan_comp_exposure.set_title("Plan Exposure")
# ax_plan_comp.set_xlabel("x [m]")
# ax_plan_comp.set_ylabel("y [m]")
# ax_plan_comp.set_zlabel("z [m]")

ax_plan_comp_duration.scatter(plan_id,duration,marker='+', color="blue")
ax_plan_comp_rail.scatter(plan_id,railing_min,marker='v', color="green",label="Minimum Demand")
ax_plan_comp_rail.scatter(plan_id,railing_mean,marker='*', color="orange",label="Average Demand")
ax_plan_comp_rail.scatter(plan_id,railing_max,marker='p', color="red",label="Maximum Demand")
# ax_plan_comp_duration.set_xticklabels(plan_id, rotation = 45)
# ax_plan_comp_rail.set_xticklabels(plan_id, rotation = 45)
# ax_plan_comp_exposure.set_xticklabels(plan_id, rotation = 45)
ax_plan_comp_duration.set_xticklabels([],visible = False)
ax_plan_comp_rail.set_xticklabels([],visible = False)
ax_plan_comp_exposure.scatter(plan_id,exposure_min,marker='v', color="green",label="Minimum Exposure")
ax_plan_comp_exposure.scatter(plan_id,exposure_mean, marker='*',color="orange",label="Average Exposure")
ax_plan_comp_exposure.scatter(plan_id,exposure_max, marker='p', color="red", label="Maximum Exposure")
ax_plan_comp_exposure.legend(loc='upper right')
ax_plan_comp_rail.legend(loc='upper right')
ax_plan_comp_duration.sharex(ax_plan_comp_exposure)
ax_plan_comp_rail.sharex(ax_plan_comp_exposure)
ax_plan_comp_rail.grid()
ax_plan_comp_duration.grid()
ax_plan_comp_exposure.grid()
ax_plan_comp_exposure.set_xticklabels(plan_id, rotation = 90)
fig_plan_comp.subplots_adjust(wspace=0, hspace=0.02)
# fig_plan_comp.tight_layout()
plt.show()
```
